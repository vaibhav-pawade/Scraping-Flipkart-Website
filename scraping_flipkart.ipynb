```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
```

```python
my_url='https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'

#for setting up connection and loading the url page
uClient=uReq(my_url)

#reading page and storing in page_html
page_html=uClient.read()

#closing connection
uClient.close()

page_soup=soup(page_html,"html.parser")
#passing html page and finding div
containers=page_soup.findAll("div",{"class":"_3O0U0u"})

#printing length of container(products on that page)
print(len(containers))

#bring html in structured format
#print(soup.prettify(container[0]))

container=containers[0]

#printing inside container
print(container.div.img["alt"])

#getting price container
price=page_soup.findAll("div",{"class":"_1vC4OE _2rQ-NK"})
print(price[0].text)

#getting ratings
rating=page_soup.findAll("div",{"class":"hGSR34"})
print(rating[0].text)

filename='products.csv'
f=open(filename,"w",encoding='utf-8')

headers="Product_name,Price,Rating\n"
f.write(headers)

for container in containers:
    product_name=container.div.img["alt"]
    product=product_name
    price_container=container.findAll("div",{"class":"_1vC4OE _2rQ-NK"})
    price=price_container[0].text
    rating_container=container.findAll("div",{"class":"hGSR34"})
    rating=rating_container[0].text
    
    f.write(product+","+price+","+rating+"\n")
    print("Product name : "+product)
    print("Price : "+price)
    print("Rating : "+rating)
f.close()
```
